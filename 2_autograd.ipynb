```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)

w = torch.tensor(1.0, requires_grad=True)

for epoch in range(100):

  # forward pass
  y_hat = w * x
  loss = (y_hat - y)**2

  # backward pass
  loss.backward()

  # SGD
  with torch.no_grad():
    w -= 0.1 * w.grad

  w.grad.zero_()

  print(f'epoch:{epoch}, loss:{loss}, y_hat:{y_hat}, dl/dw:{w.grad}')

print("weight:",w.item())
```
