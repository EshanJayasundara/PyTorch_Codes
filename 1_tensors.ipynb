Elementwise operations

```python
# 2x2 tensors with random initialized integer values in between 1 and 10
x = torch.randint(1,10,(2,2))
y = torch.randint(1,10,(2,2))
print(x)
print(y)

# pairwise addition
print(x+y) # print(torch.add(x,y))

# pairwise subtraction
print(x-y) # print(torch.sub(x,y))

# pairwise multiplication
print(x*y) # print(torch.mul(x,y))

# pairwise division
print(x/y) # print(torch.div(x,y))
```

Add x, y and put the result on y

```python
# in-place operation is common for all all operations

# example for in-place addition
y.add_(x)
print(y)
```

Tensor -> Numpy

```python
a = torch.ones(2, 3)
print(a)

b = a.numpy()
print(b)

b[0,0] = 0
print(a)
```

Numpy -> Tensor

```python
a = np.ones((2, 3))
print(a)

b = torch.from_numpy(a)
print(b)

b[0,0] = 0
print(a)
```

Using CUDA

```python
if torch.cuda.is_available():
  device = torch.device('cuda')
else:
  device = torch.device('cpu')

x = torch.ones(5, device=device)
print(x.device)
```

Numpy doesn't support CUDA. So, convert back to CPU

```python
x = x.to('cpu')
print(x.device)
```

```python
x = x.numpy()
print(type(x))
```
